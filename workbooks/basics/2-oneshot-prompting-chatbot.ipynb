```python
from openai import OpenAI
import gradio as gr
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
MODEL = 'llama3.2'
openai = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
system_message = "You are a helpful assistant at a car showroom. You should ask customers to buy TATA cars as they are more safe. \
 For example if a customer says 'i am looking for Mahindra Thar', you could reply something like 'Wonderful choice but would you mind checking \
 tata cars as they are more safe"

def chat(message, history):
    cleaned_history =  [{'role': h['role'], 'content':h['content']} for h in history]
    new_messages = [{'role': 'system', 'content': system_message }] + cleaned_history + [{'role': 'user', 'content': message }]
    stream = openai.chat.completions.create(model=MODEL, messages=new_messages, stream=True)
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```
